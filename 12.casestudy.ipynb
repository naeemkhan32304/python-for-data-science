```python
# saving datafram into csv file
kashti.to_csv("kashti.csv")
```

```python
# basic statistic or summary
kashti.describe()
```

```python
kashti.head()
```

```python
# droping few colums and make a new data set
new_kashti = kashti.drop(['deck','alone'], axis=1)
new_kashti.head()
```

```python
# Same used
kashti.drop(['deck','alone'], axis=1).head()
```
